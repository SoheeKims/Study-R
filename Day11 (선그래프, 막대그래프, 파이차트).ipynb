# 선그래프

```r
# 3종류 data의 선그래프 plot을 통해 base, line 그리고... layer 겹침 

abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
options(warn = -1) #노트북 1대당 한 번 
par(family = "AppleGothic")

plot(abc,
    type = "o",
    col = "red",
    ylim = c(0, 400), # x 축과 범위 맞추기 위해서 
    axes = F, #x,y의 좌표 Frame을 제거 
    ann = F #x,y의 라벨 제거 (index, abc) 
    )

# x축과 x축의 라벨을 설정 axis(x축은 1, 데이터 범위, lab 라벨) 
axis(1, at=1:5, lab = c("서울", "대전", "대구", "광주", "원주"))

# y축과 y축의 라벨을 설정 (y축은 2,) 
axis(2, ylim = c(0, 400))

# main title 지정 
title(main = "과일 판매량", col.main = "red")

# x축 title 지정 
title(xlab = "지역", col.lab = "blue")

# y축 title 지정 
title(ylab = "판매량", col.lab = "blue")

# def 그리기 
lines(def, 
      type = "o", 
      pch = 22, # pch 22는 네모
      col = "green", 
      lty = 2 # lty 2는 점선 
      )

# ghi 그리기 
lines(ghi, 
     type = "o",
     pch = 22, 
     col= "blue",
     lty = 2
     )

# 범례 (각 색깔이 무엇을 뜻하는지) 
legend(4, 400, # 어디에 그릴지 정하기
      c("야구장","축구장","해변가"), # 색깔 지정한 순서대로 
      col = c("red","green","blue"),
      pch = c(21, 22, 22), # 동그라미, 네모, 네모
      lty = c(1, 2, 2), 
      cex = 0.8 # 1 기본값
       )
```

# 대선자료 시각화

```r
election <- read.csv("../Data/election_result_ansi.csv", fileEncoding = "euc-kr")
head(election) # pop 투표한 사람들 집계. moon+hong+ahn < pop 
```

```r
tail(election)
```

```r
str(election)  # 시군은 중복되어 있는 이름이 있기 때문에 250보다 적은 227개 (ex. 중구)
```

```r
# electionSum = moon + hong + ahn 구해서 
# elcetion에 컬럼 추가하기 

#head(election)
election$electionSum <- election$moon + election$hong + election$ahn #sum(moon,hong,ahn)을 하면 각 후보를 전부 다 더한 값이 나오게 됨.
#election$electionSum <- apply(election[,5:7],1,sum) #5~7열만 가로방향 합계를 구한다

head(election)
```

```r
# electionRatio 컬럼 만들기: electionSum이 pop의 몇 percent? 
election$electionRatio <- election$electionSum / election$pop *100
head(election)
```

```r
#rm(election$Sum)
```

```r
# electionRatio가 제일 높은 지역은? 
subset(election, electionRatio == max(election$electionRatio)) #subset 조건 
```

```r
library(doBy)
orderBy( ~ -electionRatio, election)[1,]
```

```r
# 90% 이상 지역 추출.
unique(subset(election, electionRatio >= 90)[,2])
```
