# 제어문(조건문, conditional)

```r
x <- runif(1)*100

if(x > 5){
    print("x > 5")
    print(x)
}
print("---End---")
```

```r
x <- runif(1)*100

if(x > 50){
    print("x > 50")
    print(x)
}else if(x>25){
    print("x > 25")
    print(x)
}else{
    print("x <= 25")
    print(x)
}
print("---End---")
```

```r
# 홀수 짝수 구분

a<-7

if(a %% 2 ==0){
    #print("짝수") # paste로도 사용 가능
    cat(a, "짝수 입니다.")
}else{
    #print("홀수")
    cat(a, "홀수 입니다.")
}
```

```r
# 홀짝수 구분 : R함수
a <- c(8,10,11)
ifelse(a %% 2==0, "짝수","홀수")
ifelse(a %% 2==0, paste(a,":짝수"),paste(a,":홀수"))
```

```r
# 점수를 입력받아 학점으로 변경해 출력하기
num1 <- 89

if(num1 >= 90){
    print("A학점")
}else if(num1 >= 80){
    print("B학점")
}else if(num1 >= 70){
    print("C학점")
}else if(num1 >= 60){
    print("D학점")
}else {
    print("F학점")
}
```

```r
# q1컬럼의 Data값이 3보다 크면 data의 제곱근을 출력하고, 아니면 원래의 Data 출력하기
#ifelse(excel$q1 > 3, sqrt(excel$q1), excel$q1)
q1 <- excel$q1
ifelse(excel$q1>=3, sqrt(q1), q1)
```

```r
# 위에서 계산한 data를 excel의 q6 컬럼으로 만들어 넣기
excel$q6 <- ifelse(excel$q1>=3, sqrt(q1), q1)
head(excel)
```

```r
# 1~10까지의 합계

sum1 <- 0
for(i in 1:10){
    sum1 <- sum1+i
}
print(sum1)
```

```r
# 1부터 100까지의 짝수 합 구하기
sum1 <- 0
for(i in 1:100){
    if(i %% 2 ==0){
        sum1 <- sum1+i
    }
}
print(sum1)
```

```r
# 1부터 100까지의 짝수의 합과 홀수의 합 구하기
sum1 <- 0
sum2 <- 0
for(i in 1:100){
    if(i %% 2 == 0){
        sum1<-sum1+i
    }else{
        sum2<-sum2+i
    }
}
print(sum1)
print(sum2)

cat("1부터 100까지의 짝수의 합은",sum1,"이고 홀수의 합은",sum2,"입니다.\n")
```

```r
# excel의 q3 컬럼 데이터의 홀수의 합과 짝수의 합 구하기
head(excel)
sum1<-0
sum2<-0
for(i in excel$q3){
    if(i %% 2!=0){
       sum1<-sum1+i 
    }else{
        sum2<-sum2+i
    }
}
print(sum1)
print(sum2)
cat("excel의 q3칼럼의 홀수 합은",sum1,"이고, 짝수 합은",sum2,"입니다.\n")
```

```r
# excel의 q3컬럼 데이터 중 3의 배수의 합계를 구하시오.
sum1<-0
for(i in excel$q3){
    if(i %% 3 == 0){
       sum1<-sum1+i 
    }
}
print(sum1)
```
