# 벡터 자료 처리

```r
x <- c(1, 3, 5, 7)
y <- c(3, 5)
```

```r
# 합집합
union(x, y)
```

```r
# 교집합
intersect(x, y)
```

```r
# 차집합
setdiff(x, y)
```

```r
# 벡터 자료 정렬

#runif(10)
#sample(1:10, 2)
#sample(1:45, 6)
sample(1:10, 10) # 뒤섞기, shuffle, 무작위 추출
sort(sample(1:10, 10)) # 1차원 데이터(벡터)만 정렬 가능.
```

```r
x <- rep(c("a", "b", "c"), 4);x
```

```r
# 벡터에 a가 있으면 1을 보여준다.
match(x, "a")

# 벡터에 a와 b가 있으면 1과 2를 보여준다.
match(x, c("a", "b"))

# 벡터에 a가 몇개 있는지 출력
table(match(x, "a"))
sum(match(x,"a"), na.rm=T)

# 벡터에 a와 b가 각각 몇개인지 출력
table(match(x, c("a","b")))
```

```r
name <- c("Apple", "Computer", "Samsung", "Communication")

# name에서 Co를 갖고있는 단어의 번지수
grep("Co", name)

# name에서 Co를 갖고있는 단어 출력
name[grep("Co",name)]
name[c(2,4)]

# name에서 om을 갖고있는 단어의 번지수
grep("om", name)

# name에서 om을 갖고있는 단어 출력
name[grep("om", name)]
```
