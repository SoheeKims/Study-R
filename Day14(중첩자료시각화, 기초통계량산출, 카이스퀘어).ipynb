## 확률 밀도 함수 
- 확률 변수의 분포를 나타내는 함수로, 확률 변수가 포함될 학률을 의미한다.
- 확률 밀도이기 대문에 값 자체가 확률을 의미하지는 않는다. 
- 기본 평균은 0이고 최대값은 0에서 발생한다. 

```r
dnorm(0)
```

```r
x <- seq(-4, 4, 0.1)
y <- dnorm(x, mean=0, sd=1)

plot(x,y,
    type = "l" #선모양
    )
```

```r
# 누적 분포 함수 
# -2~2사이가 데이터 가장 많이 쌓임을 확인. 
x <- seq(-4, 4, 0.1)
y <- pnorm(x, mean=0, sd=1)
plot(x,y,
    type= "l")
```

# 수학 함수의 시각화
y = x^2

```r
f <- function(x){
    y = x^2
    return(y)
    
}
```

```r
x <- 1:3
f(x) #x를 제곱한 함수 
```

```r
x <- seq(-5, 5, 0.1)
plot(x, f(x),
    type = "l")
```

```r
# y = e^(-x) -4x^2 + x + 3|x| -3
```

```r
f <- function(x){
    y = exp(-x) - 4* x^2 + x + 3 * abs(x)-3
    return(y)
}
```

```r
plot(x, f(x),
    type = "l"
    )
```

```r
# y = x^2 + 2x + 3
```

```r
f <- function(x){
    y = x^2 + 2*x + 3
    return(y)
}
```

```r
plot(x, f(x),
    type= "l"
    )
```

# 기초 통계량 산출 

```r
cafe <- read.csv("../Data/coffee.csv")
head(cafe) 
#day.cod(요일)
```

```r
# 빈도수 3잔 팔린 날 1번, 5잔 팔린 날 2번 
table(cafe$Coffee)
```

```r
# 범위 빈도수 
coffe_range <- seq(0, 60, 10)
table(cut(cafe$Coffee, breaks = coffe_range, right=F)) 
# (0,10]=0~10 (10,20]=11~20 ... (50,60]=51~60 그렇기 때문에 right=F 써야함-> [50, 60)=50~59
```

```r
# 줄기-잎 그림 (범위 내 분포가 어느 정도인지 알고 싶음)

stem(cafe$Coffee) #10단위가 6개 있는데 10,11,11,13,14,18임 
```

```r
# 변동계수(빈번하게 발생하는 게 무엇인지)
# coffee와 쥬스의 평균과 표준편차
ca <- cafe$Coffee
ca.m <- mean(ca)
ca.sd <- sd(ca)
paste("커피 평균 판매량:", round(ca.m,1), "  커피 표준편차 :", round(ca.sd,2))
```

```r
ju <- cafe$Fruit.Cup
ju.m <- mean(ju)
ju.sd <- sd(ju)
paste("쥬스 평균 판매량:", round(ju.m,1), "  쥬스 표준편차 :", round(ju.sd,2))
```

```r
# 커피 판매량의 변동계수 -> 커피가 더 많이 팔림(변동계수 낮을 수록 빈번하게 발생하는 것) 
paste("커피 변동계수 :", ca.sd / ca.m)

# 쥬스 판매량의 변동계수
paste("쥬스 변동계수 :", ju.sd / ju.m)
```

```r
sort(cafe$Fruit.Cup) #한 잔도 안팔린 날이 보임 (정렬하니까)
sort(cafe$Coffee)
```

# 가설 검정 

## 카이제곱 검정 (카이스퀘어)
: 두 범주형 변수(성별, 학년 등)가 서로 상관이 있는지 판단하는 통계적 방법 

# child별 장난감 보유 현황 
- 귀무가설 : child1과 child2의 장난감 보유수에 대한 만족도는 차이가 없다.  
- 대립가설 : child1과 child2의 장난감 보유수에 대한 만족도는 차이가 있다.

```r
# child별 장난감 보유 현황
child1 <- c(5, 11, 1)
child2 <- c(4, 7, 3)
```

```r
# 분석을 위해 데이터 합치기 
Toy <- cbind(child1, child2)
Toy
```

```r
# 행의 이름 
rownames(Toy) <- c("car","truck","doll")
Toy
```

```r
options(warn =0) # warning 일부러 만듦(아까 이미 = -1 해놔서 안뜨기 때문에)
```

```r
# 카이제곱 검정
chisq.test(Toy)
```

```r
# p-value = 0.4219 > 0.05(95% 신뢰구간) --> 귀무가설 채택 
# child1과 child2의 데이터는 통계적으로 큰 차이가 없다. 

# p-value는 0에 가까울 수록 좋다. 미리 정해진 유의수준(일반적으로 0.05)보다 작으면 
# 대립가설을 채택하고, 크면 귀무가설을 채택함. 
```

```r
# warning 메세지 : 카이제곱의 근사값이 정확하지 않을 수도 있다. “Chi-squared approximation may be incorrect”
# 데이터량이 적은 경우 발생 
# 적은 경우 -> 피셔 검정: 표본수가 적거나 데이터의 분포가 치우친 경우 사용하는 검정 
```

```r
# 피셔 검정 (cbind해서 matrix만들어서 matrix만 넣어야함)
fisher.test(Toy)
```

```r
# p-value = 0.5165 > 0.05 --> 귀무가설 채택
```
